```python
# Package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
sns.set_style('darkgrid')
import yfinance as yf
import datetime as dt
```

```python
# Check if the APPle returns for the last year has a normal distribution

start_date = dt.datetime.today()- dt.timedelta(days = 365)
end_date = dt.datetime.today()

Apple_stock= yf.download('AAPL', start = start_date, end=end_date)

returns = Apple_stock['Close']/Apple_stock['Close'].shift(1)
logreturns = np.log(returns.dropna())['AAPL'].values
plt.figure(figsize = (10,6))
plt.plot(Apple_stock['Close'])
plt.title('AAPL stock data for a year', size = 25)
plt.show()
p_Apple=stats.normaltest(logreturns)[1]
print(f"Apple log return distribution: p-value = {p_Apple:.4f}")
if p_Apple< 0.05:
    print(" Statistically significant evidence that the data is NOT normally distributed.")
else:
    print(" No statistically significant evidence against normality.")
```

## I check if there is any month in the past 6 months such that the Apple stock has a normal distribution

```python
for i in range(5):
    start_date = dt.datetime.today() - dt.timedelta(days = (i+1)*30)
    end_date = dt.datetime.today()
    stock_i = yf.download('AAPL', start = start_date, end = end_date)
    returns_i = stock_i['Close']/stock_i['Close'].shift(1)
    logreturns_i = np.log(returns_i.dropna())['AAPL'].values
    plt.figure(figsize = (12,8))
    plt.plot(stock_i['Close'])
    plt.title('AAPL in month', size = 25)
    plt.show()
    p_Apple_i=stats.normaltest(logreturns_i)[1]
    print(f"Apple log return distribution: p-value = {p_Apple_i:.4f}")
    if p_Apple_i < 0.05:
        print(" Statistically significant evidence that the data is NOT normally distributed.")
    else:
        print(" No statistically significant evidence against normality.")
        break
print('--'*40) 
```

## Check if removing some extremal data gives us normal distribution for the Apple stock.

```python
lastmonth_index = int(np.floor(len(logreturns)/12)*11) #choice of numbers 11 and 12: I am dividing data in 12 equal parts then and want to keep only the last part.

Apple_logreturns_historical = logreturns[lastmonth_index:]
Apple_historical = Apple_stock['Close'].iloc[lastmonth_index:]

# Plot
plt.figure(figsize=(12, 8))
plt.plot(Apple_historical)
plt.title('Apple (Last month)', fontsize=25)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()

plt.figure(figsize=(12,8))
stats.probplot(Apple_logreturns_historical, plot = plt)
plt.title('Apple Log Returns (Last Half) Probability Plot', size = '25')
plt.show()

p_apple_historical = stats.normaltest(Apple_logreturns_historical)[1]

print(f"Apple log return (Last Half) distribution: p-value = {p_apple_historical:.4f}")

if p_apple_historical < 0.05:
    print(" Statistically significant evidence that the data is NOT normally distributed.")
else:
    print(" No statistically significant evidence against normality.")
```

## I  test if the stocks chosen in the first mini project are normally distributed.

```python
#one year of data
start_date = dt.datetime.today() - dt.timedelta(days = 365)
end_date = dt.datetime.today()

#Centuries Community Inc. stock data, isolate the logreturn data as a numpy array.
CCS_stock = yf.download('CCS', start = start_date, end = end_date)
CCS_returns = CCS_stock['Close']/CCS_stock['Close'].shift(1)
CCS_logreturns = np.log(CCS_returns.dropna())['CCS'].values

#Conduent Inc. stock data, isolate the logreturn data as a numpy array.
CNDT_stock = yf.download('CNDT', start = start_date, end = end_date)
CNDT_returns = CNDT_stock['Close']/CNDT_stock['Close'].shift(1)
CNDT_logreturns = np.log(CNDT_returns.dropna())['CNDT'].values

#Dole PLC stock data, isolate the logreturn data as a numpy array.
DOLE_stock = yf.download('DOLE', start = start_date, end = end_date)
DOLE_returns = DOLE_stock['Close']/DOLE_stock['Close'].shift(1)
DOLE_logreturns = np.log(DOLE_returns.dropna())['DOLE'].values

#Fresh Del Monte stock data, isolate the logreturn data as a numpy array.
FDP_stock = yf.download('FDP', start = start_date, end = end_date)
FDP_returns = FDP_stock['Close']/FDP_stock['Close'].shift(1)
FDP_logreturns = np.log(FDP_returns.dropna())['FDP'].values

#NVIDIA stock data, isolate the logreturn data as a numpy array.
NVDA_stock = yf.download('NVDA', start = start_date, end = end_date)
NVDA_returns = NVDA_stock['Close']/NVDA_stock['Close'].shift(1)
NVDA_logreturns = np.log(NVDA_returns.dropna())['NVDA'].values

#Quantum Computing Inc. index data, isolate the logreturn data as a numpy array.
QUBT_stock = yf.download('QUBT', start = start_date, end = end_date)
QUBT_returns = QUBT_stock['Close']/QUBT_stock['Close'].shift(1)
QUBT_logreturns = np.log(QUBT_returns.dropna())['QUBT'].values
```

```python
# Create 1x2 grid of subplots
fig, axes = plt.subplots(1, 6, figsize=(20, 4))

# Plot Centuries community Inc.. log returns
stats.probplot(CCS_logreturns, dist="norm", plot=axes[0])
axes[0].set_title("CCS Log Returns Probability Plot", fontsize=10)
axes[0].grid(True)

# Plot Conduent Inc. log returns
stats.probplot(CNDT_logreturns, dist="norm", plot=axes[1])
axes[1].set_title("CNDT Log Returns Probability Plot", fontsize=10)
axes[1].grid(True)

# Plot Dole PLC log returns
stats.probplot(DOLE_logreturns, dist="norm", plot=axes[2])
axes[2].set_title("DOLE Log Returns Probability Plot", fontsize=10)
axes[2].grid(True)


# Plot Fresh Del Monte log returns
stats.probplot(FDP_logreturns, dist="norm", plot=axes[3])
axes[3].set_title("FDP Log Returns Probability Plot", fontsize=10)
axes[3].grid(True)

# Plot Nvidia log returns
stats.probplot(NVDA_logreturns, dist="norm", plot=axes[4])
axes[4].set_title("NVDA Log Returns Probability Plot", fontsize=10)
axes[4].grid(True)


# Plot quantum computing Inc. log returns
stats.probplot(QUBT_logreturns, dist="norm", plot=axes[5])
axes[5].set_title("QUBT Log Returns Probability Plot", fontsize=10)
axes[5].grid(True)

plt.show()
```

```python
#Collect p-values of normality tests
p_ccs=stats.normaltest(CCS_logreturns)[1]
p_cndt=stats.normaltest(CNDT_logreturns)[1]
p_dole=stats.normaltest(DOLE_logreturns)[1]
p_fdp=stats.normaltest(FDP_logreturns)[1]
p_nvda=stats.normaltest(NVDA_logreturns)[1]
p_qubt = stats.normaltest(QUBT_logreturns)[1]

# Perform the D'Agostino and Pearsons test for Normality on the data sets, 
#We'll use p-value .05 as statistical evidence the data is not collected from a normal distribution


#dictionary of p-values
p_value_dictionary = {'Centuries Company Inc.':p_ccs, 'Conduent Inc.':p_cndt, 'Dole PLC': p_dole, 'Fresh Del Monte': p_fdp, 'NVIDIA': p_nvda, 'Quantum Computing Inc.': p_qubt}

print("Normality Test Results (D'Agostino and Pearson's Test):")
print('-------------------------'*3)
for name in p_value_dictionary:
    p = p_value_dictionary[name]
    print(f"{name}: p-value = {p}")
    if p < 0.05:
        print(" Statistically significant evidence that the data is NOT normally distributed.")
    else:
        print(" No statistically significant evidence against normality.")
    print('-------------------------')
    print('-------------------------')
```
